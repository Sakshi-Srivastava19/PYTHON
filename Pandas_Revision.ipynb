#Series

It is a 1d labelled data that will store data of same or different data type

#DataFrame

It is a 2d labelled data that will store data in rows and column.

```python
import pandas  as pd # Alias
```

```python
# How to create a series that will store my labelled data in 1d format
data= [ 11 , 12 ,"Apple" , " B" , 55.55 , "bANANA"]
s1 = pd.Series(data)
print(s1)
```

```python
data= [ 11 , 12 ,"Apple" , " B" , 55.55 , "bANANA"]
print(data)
```

```python
# Series with int and float data
data= [ 11 , 12 , 55.55 , 999.99]
s1 = pd.Series(data)
print(s1)
```

```python
# Series with int data
data= [ 11 , 12 , 55 , 999]
s1 = pd.Series(data)
print(s1)
```

```python
# Series with int and string/object data
data= [ 11 , 12 , 55 , 999 ,"A" , "Apple"]
s1 = pd.Series(data)
print(s1)
```

# How to remove this dtype=

```python
data= [ 11 , 12 , 55 , 999 ,"A" , "Apple"]
s1 = pd.Series(data)
print(s1.to_string(dtype=False))
```

# How to remove the default index as label

```python
data= [ 11 , 12 , 55 , 999 ,"A" , "Apple"]
s1 = pd.Series(data)
print(s1.to_string(index=False))
```

# How to add my own labels in place of by default labels

```python
data= [ 11 , 12 , 55 , 999 ,"A" , "Apple"]
s1 = pd.Series(data, index = ["First","Second","3rd","d","e","f"])
print(s1)
```

#How to access the value in series by using Indexing operation

```python
# I want to access the value present at index 4
print(s1[4])

print(s1[5])
```

```python
data= [ 11 , 12 , 55 , 999 ,"A" , "Apple"]
s1 = pd.Series(data, index = ["First","Second","3rd","d","e","f"])
print(s1)
```

# Can a use some other data type to create a series - YES

```python
# Create a series with list
data= [ 11 , 12 , 55 , 999 ,"A" , "Apple"]
s1 = pd.Series(data)
s1
```

```python
# Create a series with tuple
data= (11 , 12 , 55 , 999 ,"A" , "Apple")
s1 = pd.Series(data)
s1
```

```python
# Create a series with set
data= {11 , 12 , 55 , 999 ,"A" , "Apple"}
s1 = pd.Series(data)
s1
```

```python
# Create a series with dictionary  ( keys will act as labels)
data= {"name":"Python" , "DOB":1992}
s1 = pd.Series(data)
s1
```

# Slicing operation in series

```python
data= [ 11 , 12 , 55 , 999 ,"A" , "Apple"]
s1 = pd.Series(data)
s1
```

```python
# I want to access 12 to "A"
print(s1[ 1:5 ])    # start index : end index + 1
```

```python
data= [ 11 , 12 , 55 , 999 ,"A" , "Apple"]
s1 = pd.Series(data, index = ["First","Second","3rd","d","e","f"])
print(s1)
```

```python
# I want to access from 12 to "A"
print(s1[ "Second": "e" ])    # start index : end index
```

```python
data= [ 11 , 12 , 55 , 999 ,"A" , "Apple"]
s1 = pd.Series(data, index = ["1","2","3","4","5","6"])
print(s1)
```

```python
# I want to access from 11 to 999
print(s1[ "1":"4" ])
```
