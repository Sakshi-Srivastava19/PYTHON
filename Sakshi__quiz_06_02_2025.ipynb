# **6Question: A company tracks the number of hours worked by employees in a week. The data for 10 employees is as follows: 40, 42, 38, 45, 50, 55, 30, 60, 48, 120. Identify any outliers in the data using the IQR method.**

```python
data = [40, 42, 38, 45, 50, 55, 30, 60, 48, 120]

Data=pd.DataFrame()
Data["data"]=data
sns.boxplot(Data["data"])
plt.show()
```

```python
Q1 = Data["data"].quantile(0.25)
Q3 = Data["data"].quantile(0.75)

IQR = Q3-Q1
ll=Q1-(IQR*1.5)
ul=Q3+(IQR*1.5)

new_data=Data[(Data["data"]>=ll)&(Data["data"]<=ul)]

print(new_data)
Data
```

# **8Question:A teacher claims that the average height of the students in her class is 160 cm. A random sample of 10 students has the following heights: 158, 162, 155, 167, 159, 160, 164, 157, 163, 161. Test at a 5% significance level whether the average height is different from 160 cm using a one-sample t-test.**

```python
height = [158, 162, 155, 167, 159, 160, 164, 157, 163, 161]

mean=np.mean(height)

# H0 = average height is 160cm
# H1 = average height is not 160cm

from scipy.stats import ttest_1samp

t_stat , p_value=ttest_1samp(height,mean)

print(f"t_stat : {t_stat}, p_value :{p_value}")

if p_value < 0.05:
       print("we accpet H1,average height is not 160cm")
else:
  print("we accept H0, average height is 160cm ")
```

# **9Question: A consumer wants to compare the average prices of smartphones from three different brands: Brand X, Brand Y, and Brand Z. The prices of 5 smartphones from each brand are as follows:**
 Brand X Prices: 300, 350, 400, 450, 500
 Brand Y Prices: 280, 320, 360, 410, 470
 Brand Z Prices: 310, 340, 380, 420, 480
# The consumer wants to test at a 5% significance level whether there is a significant difference in the average prices between the three brands using a One-Way ANOVA test.**

```python
Brand_X_Prices= [300, 350, 400, 450, 500 ]
Brand_Y_Prices= [280, 320, 360, 410, 470]
Brand_Z_Prices= [310, 340, 380, 420, 480]

# creating null and alternative hypothesis :

# H0 = there is no significant difference in average prices.
# H1 = there is a significant difference in average prices.


import scipy.stats as stats

f_stats,p_value=stats.f_oneway(Brand_X_Prices,Brand_Y_Prices,Brand_Z_Prices)


print(f"f_stats : {f_stats}, p_value :{p_value}")


if p_value < 0.05:
      print("we accept the H1 , there is a significant difference in average prices")
else:
  print("we accpet the H0 , there is no significant difference in average prices")
```
