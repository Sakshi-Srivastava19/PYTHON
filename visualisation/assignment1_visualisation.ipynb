**1. Start off by importing the cars.csv file in the jupyter notebook**


```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\hp\Downloads\cars.csv")
df.head()
```

**2. Generate a line plot graph for the column ‘model’ and ‘hp’:<br>a. Map the ‘model’ column on the x-axis
<br>b. Map the ‘hp’ column on the y-axis
<br>c. Provide the x-axis label as Models of the cars
<br>d. Provide the y-axis label as Horse-Power of Cars
<br>e. Set the title as Model Names vs Horse**

```python
plt.figure(figsize=(10, 6))
plt.plot(df['model'], df['hp'], marker='o',markerfacecolor='red', color='b',markeredgecolor='black', markersize=8, linestyle='-', linewidth=2)
plt.title('Model Names vs Horse', fontsize=14)
plt.xlabel('Models of the cars', fontsize=12)
plt.ylabel('Horse-Power of Cars', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
