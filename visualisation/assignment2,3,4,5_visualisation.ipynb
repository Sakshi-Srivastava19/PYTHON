# Problem Statement:
**You work in XYZ Corporation as a Data Analyst. Your corporation has told you to
visualize the mtcars.csv dataset with various plots.
<br>Tasks To Be Performed:<br>1. Generate a bar plot graph for the columns ‘carbs’ and ‘gear’:<br><br>
a. Map the ‘carbs’ onto the x-axis<br>
b. Map the ‘gear’ onto the y-axis<br>
c. Provide the x-axis label as Number of carburetors<br>
d. Provide the y-axis label as Number of forward gears<br>
e. Set the title as carbs vs gear**


```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\hp\Downloads\mtcars.csv")
df.head()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['carb'], df['gear'], color='pink')
plt.title('Carbs vs Gear', fontsize=14)
plt.xlabel('Number of Carburetors', fontsize=12)
plt.ylabel('Number of Forward Gears', fontsize=12)
plt.tight_layout()
plt.show()
```

# Problem Statement:
**You work in XYZ Corporation as a Data Analyst. Your corporation has told you to
visualize the mtcars.csv dataset with various plots.<br>
Tasks To Be Performed:<br>1. Plot a pie chart for columns: ‘cyl’ and ‘model’ form the mtcars.csv data
frame.**

```python
# Count the occurrences of each 'cyl' value
cyl_counts = df['cyl'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(4, 4))
plt.pie(cyl_counts, labels=cyl_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Distribution of Number of Cylinders in Cars', fontsize=14)
plt.axis('equal')
plt.show()
```

```python
cyl_counts = df['cyl'].value_counts()
models_by_cyl = df.groupby('cyl')['model'].apply(lambda x: ', '.join(x))
plt.figure(figsize=(12, 18))
plt.pie(cyl_counts, labels=[f'{cyl} Cyl: {models_by_cyl[cyl]}' for cyl in cyl_counts.index], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Distribution of Number of Cylinders and Models', fontsize=46)
plt.axis('equal')
plt.show()
```

# Problem Statement:
**You work in XYZ Corporation as a Data Analyst. Your corporation has told you to
visualize the mtcars.csv dataset with various plots.
<br>Tasks To Be Performed:<br>1. Plot the area chart for the columns ‘am’ and ‘carb’:
a. Set the ‘am’ on the x-axis<br>
b. Set the ‘carb’ on the y-axis<br>
c. Provide the x-axis label as Transmission<br>
d. Provide the y-axis label as Number of carburetors<br>
e. Provide the title as Transmission vs Number of carbuetors.**

```python
plt.plot(df['am'], df['carb'])
plt.title('Transmission vs Number of carburetors')
plt.xlabel('Transmission')
plt.ylabel('Number of carburetors')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.fill_between(df['am'], df['carb'], color='blue', alpha=0.5)
plt.title('Transmission vs Number of Carburetors', fontsize=18)
plt.xlabel('Transmission', fontsize=12)
plt.ylabel('Number of Carburetors', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
df_grouped = df.groupby('am')['carb'].sum()
plt.figure(figsize=(10, 6))
df_grouped.plot(kind='area', color='yellow', alpha=0.5)
plt.title('Transmission vs Number of Carburetors', fontsize=14)
plt.xlabel('Transmission', fontsize=12)
plt.ylabel('Number of Carburetors', fontsize=12)
plt.tight_layout()
plt.show()
```
